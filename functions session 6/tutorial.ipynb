### Variable Scope

#### Global variable vs local variable

````python
def g(y):
    print(y)
    print(y+1)

x = 5
g(x)
print(x)
````

```python
def g(y):
    print(y)
    print(y+1)

x = 5
g(x)
print(x)
```

### What the Code Does

1. **Define a Function**: 
   The first part of the code is defining a function called `g` that takes one parameter, `y`. Inside the function, it prints the value of `y` and then prints the value of `y + 1`.

2. **Set a Variable**: 
   After the function is defined, the code sets a variable `x` to 5. This means that `x` now holds the value 5.

3. **Call the Function**: 
   Next, the code calls the function `g` and passes the value of `x` to it. This means `g(x)` is the same as `g(5)` because `x` holds the value 5.

4. **Print the Variable**: 
   Finally, the code prints the value of `x`.

### Step-by-Step Execution

- When the code runs, it first goes through the function definition `def g(y):`. The interpreter recognizes that a function is being defined and stores it in memory, but does not run it yet.

- Next, `x = 5` is executed. This line creates a variable `x` and assigns it the value 5.

- Then, the function `g(x)` is called. This means the code jumps to the function `g` and runs it. Since `x` is 5, it passes 5 into the function, so `y` becomes 5 inside the function.

- Inside the function `g`:
  - `print(y)` prints the value of `y`, which is 5.
  - `print(y+1)` prints the value of `y + 1`, which is 6.

- After the function has finished running, the code returns to the next line after `g(x)`, which is `print(x)`. This prints the value of `x`, which is still 5.

### Key Points to Notice

- When the function is called with `g(x)`, the value of `x` (which is 5) is passed to the function, and it becomes `y` inside the function. This is a local variable, meaning it only exists while the function is running.
  
- The original variable `x` outside the function remains unchanged, even after the function is done. This is why when we print `x` at the end, it still shows 5.

- In this code:
  - `y` is a local variable (only exists in the function `g`).
  - `x` is a global variable (exists outside the function).

The main idea is that changing `y` inside the function doesn't affect `x` outside the function, even though they started with the same value.

#### Concept of Function References

In Python, functions are first-class objects, which means:
1. **Functions can be passed around as arguments** to other functions.
2. **Functions can be returned from other functions**.
3. **Functions can be assigned to variables**.

When a function is assigned to a variable or returned by another function, what we are dealing with is a *function reference*. This is not calling the function but rather referring to it. 

#### Code Analysis

```python
def f():
    def x(a, b):
        return a + b
    return x

val = f()(3, 4)
val
```

#### Breakdown of above code

1. **Defining `f` and `x`**:
   - `f()` is defined, and inside it, another function `x(a, b)` is defined.
   - The function `x` takes two arguments and returns their sum.

2. **Returning a Function Reference**:
   - When `f()` is called, it doesn't execute `x(a, b)` right away. Instead, it reaches the `return x` line.
   - Here, `x` is not called; instead, a reference to the function `x` is returned. This means that `f()` provides us with the ability to access and use the `x` function outside of its local scope.

3. **Using the Function Reference**:
   - `f()` returns a reference to `x`, which is a function. Thus, `f()` can be thought of as returning `x` itself.
   - When we write `f()(3, 4)`, this expression is equivalent to taking the reference of `x` returned by `f` and then calling it with the arguments `3` and `4`.

4. **Execution Flow**:
   - Let's break down `f()(3, 4)`:
     - `f()` is called: This returns a reference to the function `x`.
     - `x(3, 4)`: Now that we have the reference to `x`, we use it to call `x` with arguments `3` and `4`.

5. **Assignment to `val`**:
   - The call `x(3, 4)` computes the sum of `3` and `4`, which is `7`.
   - This result (`7`) is then assigned to the variable `val`.

#### Visual Representation

To visualize, here’s a simplified flow:

1. `f()` → Returns reference to `x`.
2. `f()(3, 4)` → `x(3, 4)` (using the reference returned by `f()`).
3. `x(3, 4)` → Computes `3 + 4`, which is `7`.
4. `val = 7`.

#### Improved Version with Function References Emphasized

To make the function reference concept even clearer, let's add comments to the code:

```python
# Define the outer function f
def f():
    # Define the inner function x, which takes two arguments and returns their sum
    def x(a, b):
        return a + b
    # Return a reference to the inner function x
    return x

# Get the reference of the inner function x by calling f(), and then call x with arguments 3 and 4
val = f()(3, 4)  # This is equivalent to getting x reference from f() and then calling x(3, 4)

# Output the result stored in val
print(val)  # Outputs: 7
```

#### Summary from the Function Reference Perspective

- **Function Reference**: In this example, when we call `f()`, it does not execute `x` but instead returns a reference to `x`. 
- **Using the Reference**: The returned function reference is then used to call `x` with specific arguments (`3` and `4`).
- **Execution Flow**: The ability to return a function reference and then call it is what makes higher-order functions powerful. It allows for flexible and dynamic execution, where functions can generate or return other functions.

This approach of using function references is powerful in Python, enabling techniques like decorators, callbacks, and functional programming patterns. It gives you the flexibility to dynamically decide which functions to execute and when, based on the logic of your program.

#### Code

```python
def func_a():
    print('inside func_a')

def func_b(z):
    print('inside func_b')
    return z()
```

#### Breaking down the code

1. **Defining `func_a`**:
    - `func_a()` is a simple function that, when called, prints the string `'inside func_a'`.

2. **Defining `func_b`**:
    - `func_b(z)` is defined to take a single argument `z`.
    - Inside `func_b`, it prints the string `'inside func_b'`.
    - Then, it calls `z()` and returns the result of this call.

#### Key Concepts of Function References in This Code

- **Passing Function References**: 
    - Here, `z` is intended to be a function reference that is passed as an argument to `func_b`. This means that `z` should be a function that can be called within `func_b`.
  
- **Calling the Passed Function**: 
    - `z()` is called inside `func_b`. Since `z` is expected to be a function reference, `z()` executes that function.
    - By calling `z()`, `func_b` is executing whatever function was passed to it via `z`.

#### Example Usage of `func_a` and `func_b`

To understand how these functions work together, let's look at an example:

```python
# Define func_a, which just prints a message
def func_a():
    print('inside func_a')

# Define func_b, which takes a function as an argument, prints a message, and then calls that function
def func_b(z):
    print('inside func_b')
    return z()

# Pass func_a as an argument to func_b
func_b(func_a)
```

#### Step-by-Step Execution

1. **Function Definitions**:
    - `func_a()` and `func_b(z)` are defined.

2. **Calling `func_b(func_a)`**:
    - Here, `func_b` is called with `func_a` as its argument.
    - `func_a` is a function, and by passing it to `func_b`, we are passing a reference to the `func_a` function, not calling it yet.

3. **Inside `func_b`**:
    - When `func_b(func_a)` is called, `func_b` begins execution, with `z` now referencing `func_a`.
    - `func_b` prints `'inside func_b'`.

4. **Calling `z()` within `func_b`**:
    - Now, `z()` is executed. Since `z` is a reference to `func_a`, this is equivalent to calling `func_a()`.
    - This means `func_a()` is now executed, and it prints `'inside func_a'`.

#### Visual Representation of the Execution Flow

1. `func_b(func_a)` → `z` becomes a reference to `func_a`.
2. Inside `func_b` → prints `'inside func_b'`.
3. `z()` (or `func_a()`) → prints `'inside func_a'`.

#### Final Output


```
inside func_b
inside func_a
None
```

#### Summary

- **Function References**: 
    - `func_b` accepts a function reference (`z`). When calling `func_b(func_a)`, we're passing the function `func_a` as a reference, not executing it.
  
- **Using Function References**:
    - Inside `func_b`, calling `z()` is actually calling `func_a()` because `z` references `func_a`.

- **Flexibility and Dynamic Execution**:
    - This method allows you to dynamically decide which function to execute inside `func_b` by simply changing the function passed as an argument.

This pattern is very common in Python, especially in scenarios involving callbacks, decorators, and higher-order functions, where you want one function to manage the execution of other functions dynamically. It allows you to decouple the function execution from function definition and makes your code more modular and reusable.
