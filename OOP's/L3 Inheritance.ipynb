## Inheritance

### What is inheritance

Inheritance says what ever belongs to the father belongs to the sun as well.

So basically the child class can access the attributes and methods of the parent class. This increases the code reusebility.

```python
#Parent
class User:
    def __init__(self):
        self.name = "Tanzir"
        self.gender = "Male"
    def login(self):
        print("login")

#Child
class Student(User): #This is the syntax to define if a class as an student class
    def __init__(self):
        self.rollno = 100

    def enroll(self):
        print('enroll into the course')
```

```python
user = User()
s = Student()
```

```python
s.name
```

```python
s.gender
```

***Although we can see that `name` was an `attribute` of the `user class` the `Student class` was still able to access it.***

***The reason we don't put a constructor in the `Student` class is that if we did, when the `Student` class runs, it would look for its own constructor. If the `Student` class has a constructor, it won't run the constructor of its parent class. This would cause an error when trying to access parent attributes like `name` or `gender`***

***However, if we remove the constructor from the `Student` class, it will use the constructor from the parent class instead. This way, the `name` and `gender` attributes from the parent class will be set correctly, and we can access them without any issues***

```python
#Parent
class User:
    def __init__(self):
        self.name = "Tanzir"
        self.gender = "Male"
    def login(self):
        print("login")

#Child
class Student(User): #This is the syntax to define if a class as an student class

    def enroll(self):
        print('enroll into the course')
```

```python
user = User()
s = Student()
```

```python
s.name
```

```python
s.gender
```

The `SmartPhone` class is empty, while the `Phone` class contains code. Despite the `SmartPhone` class being empty, when an object of the `SmartPhone` class is created, certain outputs are still printed.


This behavior likely occurs because the `SmartPhone` class inherits from the `Phone` class. In Python, when a subclass (like `SmartPhone`) inherits from a superclass (like `Phone`), it can access and execute the methods and properties defined in the superclass. Even if the subclass doesn't define any methods or properties, it can still use those of its parent class. Thus, when an object of the `SmartPhone` class is created, it can invoke the methods from the `Phone` class, leading to the observed printed outputs.

### Method Overriding

```python
class Phone:
    def __init__(self,price,brand,camera):
        print("Inside phone constructor")
        self.__price = price
        self.brand = brand
        self.camera = camera

    def buy(self):
        print("Buying a phone")

class SmartPhone(Phone):
    
    def buy(self):
        print("Buying a smartphone")
```

```python
s = SmartPhone(20000,"Apple",13)
```

```python
s.buy()
```

***If there is Inheritance and If both the parent class and the child class have a method with the same name, the method in the child class will be called. This concept is known as method overriding.***

### Super Keyword

```python
class Phone:
    def __init__(self,price,brand,camera):
        print("Inside phone constructor")
        self.__price = price
        self.brand = brand
        self.camera = camera
    
    def buy(self):
        print("Buying a phone")

class SmartPhone(Phone):
    def buy(self):
        print("Buying a smartphone")
        # Syntex to call parent ka buy method
        super().buy()
```

```python
s = SmartPhone(10000,"apple",20)
```

```python
s.buy()
```

***First the `buy` method of the child class get's called then the `buy` method of the parent class get's called, and it gets called using super method.***

So we basically use super method to call the method from the parent class

```python
# super -. constrctor
class Phone:
    def __init__(self,price,brand,camera):
        print("Inside the phone constructor")
        self.__price = price
        self.brand = brand
        self.camera = camera

class SmartPhone(Phone):
    def __init__(self, price,brand,camera,os,ram):
        # print("Inside smartphone constructor")
        super().__init__(price,brand,camera)
        self.os = os
        self.ram = ram
        print("inside smartphone constructor")
```

```python
s = SmartPhone(20000,"Samsung", 12, "Android", 2)
s.os
```

```python

s.brand
```

### Inheritance Summary
* A class can inherit from another class.
* Inheritance improves code reusebility
* Constructor, attributes, methods get inherited to the child class.
* The parent has no access to child class
* Private property of parent are not accessible directly in child class
* Child class can override the attributes or the methods, This is called method overriding
* Super() is an inbuilt function which is used to invoke the parent class methdos and constructors

### Few Examples

```python
class Parent:
    def __init__(self, num):
        self.__num=num

    def get_num(self):
        return self.__num

class Child(Parent):
    def __init__(self,num,val):
        super().__init__(num)
        self.__val=val

    def get_val(self):
        return self.__val
```

```python
son = Child(100,200)
```

```python
son.get_num()
```

```python
son.get_val()
```

```python
class Parent:
    def __init__(self):
        self.num=100

class Child(Parent):
    def __init__(self):
        super().__init__()
        self.var=200

    def show(self):
        print(self.num)
        print(self.var)
```

```python
son = Child()
```

```python
son.show()
```

***In above problem we have num as well as var and we can notice one thing that somehow even with self.num we are able to access the value of self.num and why is that? It's because self and son are the same thing so if son can access num self can access num as well***

```python
class Parent:
    def __init__(self):
        self.__num=100

    def show(self):
        print("Parent:",self.__num)

class Child(Parent):
    def __init__(self):
        super().__init__()
        self.__var=10

    def show(self):
        print("Child:",self.__var)
```

```python
obj=Child()
```

```python
obj.show()
```

* Here we get the entire chain of grand father attributes and methods being inherited by the father and the attributes and methods of father being inherited by the son. And this chain can increase in both direction
